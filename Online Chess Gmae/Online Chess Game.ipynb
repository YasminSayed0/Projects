```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cg=pd.read_csv('chess_games.csv')
cg
```

```python
cg.info()
```

```python
pd.DataFrame({'count':cg.shape[0],
               'null%':cg.isnull().mean() *100,
               'null':cg.isnull().sum(),
               'cardinality':cg.nunique()
    
    
})
```

```python
cg.duplicated().any()
```

```python
percentage_cg = (cg['winner'].value_counts() / len(cg)) * 100
plt.figure(figsize=(10, 6))  
sns.barplot(x=percentage_cg.index, y=percentage_cg.values)

for index, value in enumerate(percentage_cg):
    plt.text(index, value + 0.5, f'{value:.2f}%',ha='center')
    
plt.xlabel('Winner')
plt.ylabel('Percentage')
plt.title('Percentage Distribution of Winner')

plt.tight_layout()
plt.show()
```

```python
purpose_counts = cg['rated'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(purpose_counts, labels=purpose_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution Of Rated')
plt.axis('equal')  
plt.show()
```

```python
percentage_cg = (cg['victory_status'].value_counts() / len(cg)) * 100
plt.figure(figsize=(10, 6))  
sns.barplot(x=percentage_cg.index, y=percentage_cg.values)

for index, value in enumerate(percentage_cg):
    plt.text(index, value + 0.5, f'{value:.2f}%',ha='center')
    
plt.xlabel('Victory Status')
plt.ylabel('Percentage')
plt.title('Percentage Distribution of Victory Status')

plt.tight_layout()
plt.show()
```

```python
black=cg[cg['winner']=='Black']
```

```python
percentage_cg = (black['opening_code'].value_counts() / len(black)) * 100
top_12 = percentage_cg.nlargest(20)

plt.figure(figsize=(11, 7))  
sns.barplot(x=top_12.index, y=top_12.values)

for index, value in enumerate(top_12):
        plt.text(index, value, f'{value:.2f}%', ha='center',va='bottom')     
plt.xlabel('Opening Code')
plt.ylabel('Percentage')
plt.title('Percentage Distribution of Top 20 Opening Code(balck)')

plt.tight_layout()
plt.show()
```

```python
percentage_cg = (cg['white_id'].value_counts() / len(cg)) * 100
top_12 = percentage_cg.nlargest(20)

plt.figure(figsize=(12, 8))  
sns.barplot(x=top_12.index, y=top_12.values)

for index, value in enumerate(top_12):
        plt.text(index, value , f'{value:.2f}%', ha='center')  # ØªØ­Ø¯ÙŠØ¯ Ù…Ø­Ø§Ø°Ø§Ø© Ø§Ù„Ù†Øµ ÙˆØªØ¯ÙˆÙŠØ±Ù‡    
plt.xlabel('White id')
plt.ylabel('Percentage')
plt.title('Percentage Distribution of Top 20(White id)')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
plt.show()
```

```python
sum_ratings_white = cg['white_rating'].sum()
sum_ratings_black = cg['black_rating'].sum()

total_sum = sum_ratings_white + sum_ratings_black

percentage_white = (sum_ratings_white / total_sum) * 100
percentage_black = (sum_ratings_black / total_sum) * 100

plt.figure(figsize=(8, 6))
sns.barplot(x=['white', 'black'], y=[sum_ratings_white, sum_ratings_black])
plt.xlabel('Column')
plt.ylabel('Sum of Ratings')
plt.title('Sum of Ratings for Each Column')

plt.text(0, sum_ratings_white, f'{percentage_white:.2f}%', ha='center', va='bottom')
plt.text(1, sum_ratings_black, f'{percentage_black:.2f}%', ha='center', va='bottom')

plt.show()
```
